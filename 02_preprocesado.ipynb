```python
# Paso 1: Subir y descomprimir archivo ZIP
from google.colab import files
import zipfile
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# Subir archivo ZIP
uploaded = files.upload()

# Descomprimir el archivo ZIP
for filename in uploaded.keys():
    if filename.endswith('.zip'):
        with zipfile.ZipFile(filename, 'r') as zip_ref:
            zip_ref.extractall('/content')

# Paso 2: Cargar el archivo train.csv
df = pd.read_csv('/content/train.csv')

# Mostrar las primeras filas
print("Primeras filas del dataset:")
display(df.head())

# Paso 3: Información general del dataset
print("\nResumen del dataset:")
display(df.info())

# Paso 4: Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
display(df.describe(include='all'))

# Paso 5: Identificar columnas categóricas y numéricas
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

print("\nColumnas categóricas:", categorical_cols)
print("Columnas numéricas:", numerical_cols)

# Paso 6: Imputar datos faltantes
# Numéricos: se imputan con la media
imputer_num = SimpleImputer(strategy='mean')
df[numerical_cols] = imputer_num.fit_transform(df[numerical_cols])

# Categóricos: se imputan con la moda (valor más frecuente)
imputer_cat = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = imputer_cat.fit_transform(df[categorical_cols])

# Paso 7: One-hot encoding para variables categóricas
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Paso 8: Normalizar variables numéricas (opcional pero recomendado para algunos modelos)
scaler = StandardScaler()
df_encoded[numerical_cols] = scaler.fit_transform(df_encoded[numerical_cols])

# Paso 9: Verificar el resultado
print("\nShape del dataset procesado:", df_encoded.shape)
print("Primeras filas del dataset preprocesado:")
display(df_encoded.head())

# Guardar el archivo preprocesado (opcional)
df_encoded.to_csv('/content/train_preprocesado.csv', index=False)
print("Archivo 'train_preprocesado.csv' guardado en /content/")
```
